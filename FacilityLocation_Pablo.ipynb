```python
ys = np.linspace(0,1, n_x)
xs = np.linspace(0,1, n_y)
```

```python
def aligntogrid(x_data, y_data, xs, ys):
    #shifts the data to the nearest grid point
    x_new_data = []
    y_new_data = []
    for x in x_data:
        
        
        absolute_difference_function = lambda list_value : abs(list_value - x)

        closest_value = min(xs, key=absolute_difference_function)
        x_new_data.append(closest_value)
        
    for y in y_data:
        
        absolute_difference_function = lambda list_value : abs(list_value - y)

        closest_value = min(ys, key=absolute_difference_function)
        y_new_data.append(closest_value)
    return x_new_data, y_new_data
```

```python
import random

random.seed(3000) ## **Pablo** Set the seed to generate the same random customers on each run. 
def samplegrid(w, h, n):
    a = [divmod(i, (h-1)) for i in random.sample(range((w-1) * (h-1)), n)]
    anewtuple = []
    for atuple in a:
        anewtuple.append([atuple[0]/(w-1), atuple[1]/(h-1)])
        
    anewtuple = np.array(anewtuple)
    return anewtuple.T[0], anewtuple.T[1]
```

Our variables $x_i$ will be each one of the red points in the image, i.e the different ATMs in the city of Münich. 
We will solve an Ising Hamiltonian whose expression is : 
$$H=\sum J_{ij}x_ix_j+\sum h_ix_i$$ 

In this case we will set the weights $h_i$ to be the sum ${\sum d_k}$ of the euclidean distance (other metrics can be taken) of each ATM to the clients (the points in blue)

The interaction terms $J_{ij}$ will be the minus the distances between ATM $i$ and $j$ ${-d_{ij}}$ so that we will favour the activation of ATMs that are far. This way, we avoid activating two ATMs that are very close.

Also we will restrict the number of ATMS that are being activated to a number K. This restriction is going to be introduced with the expression $(\sum_{i}x_i-K)^{2}$.




First, we store the coordinates of each ATM and each client, and store it in a dictionary to make things easier. Also we delete repeated items (there are a lot) 

```python
loc_atms={} # We store in dictionaries the location of ATMS and the location of clients 
loc_clients={} 


atms_l=[] # we create list to store the locations and avoid repeated items.
clients_l=[]

count=0

for i in range(len(longitudes)):

    
    if ((longitudes[i],latitudes[i]) not in atms_l):
    
        loc_atms[f'x{count}']=(longitudes[i],latitudes[i])
        
        count+=1
        
    atms_l.append((longitudes[i],latitudes[i]))
    
    
count=0
    
for i in range(len(x_customers)):
    
    if((x_customers[i],y_customers[i]) not in clients_l):
    
        loc_clients[f'c{count}']=(x_customers[i],y_customers[i])
        
        count+=1
        
    clients_l.append((x_customers[i],y_customers[i]))
    

    
```

Now we store in dictionaries the linear h  and quadratic terms J of the euclidean distances from between ATMs and customers. 

```python
def EuclideanDistance(x,y): 
    
    return np.sqrt((x[0]-y[0])**2+(x[1]-y[1])**2)

h={}
J={}


N=len(loc_atms.keys())
C=len(loc_clients.keys())



for i in range(N):
    
    dist=0
    
    for j in range(C):
        
        dist+=EuclideanDistance(loc_atms[f'x{i}'],loc_clients[f'c{j}']) 

    
    h[i]=dist # We will priorize the atms whose distance to the clients is lower
    
for i in range(N-1):
    
    for j in range(i+1,N):
        
        dist=EuclideanDistance(loc_atms[f'x{i}'],loc_atms[f'x{j}'])
        
        J[(i,j)]=-dist # We will priorize opening two atms which are far rather than two that are close
        
```

Now we will normalize the terms and assign a weight $\lambda$ to each of them. This weight will determine the importance of the constraint. 

```python
lambda0=1  ## We choose the parameters that control the importance given to the different terms
lambda1=1
        
for k in h.keys():
    
    h[k]= lambda0*(h[k]-min(h.values()))/(max(h.values())-min(h.values()))  #We normalize the data obtained on the distances
    
for k in J.keys():
    
    J[k]= lambda1*(J[k]-min(J.values()))/(max(J.values())-min(J.values()))
    
```

We will have to introduce the restriction that only 5 ATMs can be opened. The expression of the constraint can be developed in the following way: 

$$\left(\sum_{i}x_i-K\right)^{2}=(1-2K)\sum_{i}x_i+2\sum_{i,j}x_ix_j+K^{2}$$ 








```python
K=5
lambda2=2

for i in range(N):
    
    h[i]+=lambda2*(1-2*K)
    
for i in range(N-1):
    
    for j in range(i+1,N):
        
        J[(i,j)]+=2*lambda2
        
offset=K**2


```

```python
######## We solve the problem in D-WAVE 

import dimod 

bqm=dimod.BinaryQuadraticModel(h,J,'BINARY')
sampleset=dimod.ExactSolver().sample(bqm) ## This method will use an ExactSolver

```

```python
energies=sampleset.record.field(1)
solutions=sampleset.record.field(0)
min_sols=[]

min_energy=min(energies)

for i in range(len(energies)):
    
    if energies[i]==min_energy:
        
        min_sols.append(list(solutions[i]))
        
```

```python
longitudes_min=[]
latitudes_min=[]


for i in range(len(min_sols[0])):
    
    if min_sols[0][i]==1:
        
        longitudes_min.append(loc_atms[f'x{i}'][0])
        latitudes_min.append(loc_atms[f'x{i}'][1])
        
    
        
print(longitudes_min)    
    
```
